```python
import pandas as pd
```

```python
df = pd.DataFrame()
df_data = pd.read_csv('../data/fashion.csv')
```

```python
c1 = df_data[df_data['SubCategory'] == 'Topwear'].head(10)
```

```python
c2 = df_data[df_data['SubCategory'] == 'Bottomwear'].head(10)
```

```python
c3 = df_data[df_data['SubCategory'] == 'Dress'].head(10)
```

```python
c4 = df_data[(df_data['SubCategory'] == 'Topwear') & (df_data['Gender'] == 'Boys')].head(10)
```

```python
c5 = df_data[(df_data['SubCategory'] == 'Bottomwear') & (df_data['Gender'] == 'Boys')].head(10)
```

```python
c6 = df_data[(df_data['SubCategory'] == 'Shoes') & (df_data['Gender'] == 'Men')].head(10)
c6
```

```python
df = pd.concat([c1,c2,c3,c4,c5,c6])
```
